```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD


#Random Data
np.random.seed(42)
X = np.random.rand(100, 2)
y = 3 * X[:, 0] + 2 * X[:, 1] + 1 + 0.1 * np.random.randn(100)

# Build a simple linear regression model
model = Sequential()
model.add(Dense(1, input_shape=(2,)))

# Define Stochastic Gradient Descent optimizer
sgd_optimizer = SGD(learning_rate=0.01)

# Compile the model
model.compile(loss='mean_squared_error', optimizer=sgd_optimizer)

# Training loop using SGD
batch_size = 1
epochs = 50
loss_history = []

for epoch in range(epochs):
    for i in range(0, len(X), batch_size):
        X_batch = X[i:i+batch_size]
        y_batch = y[i:i+batch_size]
        loss = model.train_on_batch(X_batch, y_batch)
    loss_history.append(loss)
    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss}")

# Print the trained model's weights
print("Trained weights:")
print(model.get_weights())
```

```python
import matplotlib.pyplot as plt
# Plotting the loss history
plt.plot(range(1, epochs+1), loss_history)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```

```python
y_pred = model.predict(X).flatten()
plt.scatter(y, y_pred, color='g')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2, color="black")
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs. True Values')
plt.show();
```
